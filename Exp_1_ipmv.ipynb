```python
#Log transform
c1=255/np.log(1+np.max(img1))
c2=255/np.log(1+np.max(img2))
```

```python
#Power law transform

images = [img2]
for gamma in [0.1, 0.5, 1.5, 2.5, 10]:

	gamma_corrected = np.array(255*(img2/ 255) ** gamma, dtype = 'uint8')

	cv2.imwrite('gamma_transformed'+str(gamma)+'.jpg', gamma_corrected)
	images.append(gamma_corrected)

titles = ['Original img','gamma = 0.1','gamma = 0.5','gamma = 1.5','gamma = 2.5','gamma = 10']

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
images = [img1]
for gamma in [0.1, 0.5, 1.5, 2.5, 10]:

	gamma_corrected = np.array(255*(img1/ 255) ** gamma, dtype = 'uint8')

	cv2.imwrite('gamma_transformed'+str(gamma)+'.jpg', gamma_corrected)
	images.append(gamma_corrected)

titles = ['Original','gamma = 0.1','gamma = 0.5',
          'gamma = 1.5','gamma = 2.5','gamma = 10']

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
#Bit Plane Slicing
b0 = (img1 >> 0) & 1
b1 = (img1 >> 1) & 1
b2 = (img1 >> 2) & 1
b3 = (img1 >> 3) & 1
b4 = (img1 >> 4) & 1
b5 = (img1 >> 5) & 1
b6 = (img1 >> 6) & 1
b7 = (img1 >> 7) & 1

plt.subplots(figsize=(10,10))

plt.subplot(4, 2, 1)
plt.imshow(b0, cmap= 'gray')
plt.title('bit 0')
plt.subplot(4, 2, 2)
plt.imshow(b1, cmap= 'gray')
plt.title('bit 1')
plt.subplot(4, 2, 3)
plt.imshow(b2, cmap= 'gray')
plt.title('bit 2')
plt.subplot(4, 2, 4)
plt.imshow(b3, cmap= 'gray')
plt.title('bit 3')
plt.subplot(4, 2, 5)
plt.imshow(b4, cmap= 'gray')
plt.title('bit 4')
plt.subplot(4, 2, 6)
plt.imshow(b5, cmap= 'gray')
plt.title('bit 5')
plt.subplot(4, 2, 7)
plt.imshow(b6, cmap= 'gray')
plt.title('bit 6')
plt.subplot(4, 2, 8)
plt.imshow(b7, cmap= 'gray')
plt.title('bit 7')
```

```python
#Contrast stretching

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
[rows,cols] = gray_img.shape;
GL_without=gray_img;
GL_with=gray_img;

r1=150
r2=200
s1=255
s2=100
s3=0

for r in range(0, rows-1):
  for c in range(0, cols-1):
    if (gray_img[r,c]<r1):
      GL_without[r,c]=s1;
    else:
      if (gray_img[r,c]<r2):
        GL_without[r,c]=s2;
      else:
        GL_without[r,c]=s3;
for r in range(0, rows-1):
  for c in range(0, cols-1):
    if (gray_img[r,c]<r1):
      GL_with[r,c]=gray_img[r,c];
    else:
      if (gray_img[r,c]<r2):
        GL_with[r,c]=s2;
      else:
        GL_with[r,c]=gray_img[r,c];
plt.imshow(GL_without)
plt.show()
```
